#### Hypothesis 1: Spark engages children from one sex more than the other.

```python
# Get UES for each sex
female_UES = (ues.loc[ues['Sex'] == "Female"]["UES"])
male_UES = (ues.loc[ues['Sex'] == "Male"]["UES"])

print("\nFemale sample:\nMean = " + str(female_UES.mean()) + "\nStandard Error = " + str(np.std(female_UES)))
print("\nMale sample:\nMean = " + str(male_UES.mean()) + "\nStandard Error = " + str(np.std(male_UES)))

# Plot
plt.boxplot([female_UES, male_UES], labels = ["Female","Male"], boxprops={"color":"Blue"}, medianprops={"color":"Red"}, vert=False)
plt.title('UES Scores per Sex')
plt.ylabel('Sex')
plt.xlabel('UES Score with Median (Red)')
plt.xlim([1,5])

# Save Figure
path = "figures/UES/UES_boxplot_perSex.png"
plt.savefig(path, dpi=300, facecolor='w', edgecolor='w',
        orientation='landscape', format="png",
        transparent=False, bbox_inches="tight", pad_inches=0.1, metadata=None)

plt.show()
```

```python
# Analyze the difference between the two groups using independent samples t-test

t, p = stats.ttest_ind(female_UES,male_UES)

print("t-statistic: " + str(t))
print("p-value: " + str(p))
```

### Interpretation


H0: There's **no difference** in engagement between the sexes

We can not reject H0 with a 95% confidence as **p > .65**. We therefore have to assume that **children from both genders felt similiarly engaged.**

### Data Analysis

#### Hypothesis 4: Spark leads to strong intention of its players to engage in further play sessions.

```python
# Set relevant mean
# Find the normal distributed mean
# mean_norm = statistics.mean(range(1,8))
mean_norm = 5

# Analyze the significance of the result with a 1-sample t-test
t, p = stats.ttest_1samp(bis["BIS"], mean_norm)

print("t-statistic: " + str(t))
print("p-value: " + str(p))
```

### Interpretation

H0: Spark **does not lead to a stronger intention** of its players to engage in further play sessions.

We can reject H0 at a confidence of .95 (p < .05). We can therefore assume that **Spark leads to a stronger intention to replay**.

### Data Import and Pre-processing (Detailed)

```python
# DO NOT RE-EXECUTE!
# NEEDS TO BE EXECUTE ONLY ON FIRST USE, IF PREPROCESSED DATA IS NOT AVAILABLE

# Fetch starting time for each trial
start_times = []
for trial in trials_pos:
    time = datetime.datetime.strptime(trial.loc[0, "Time"], '%Y/%m/%d %H:%M:%S.%f')
    start_times.append(time)
    
# Define meaningful events
meaningful_events = ["WhiteHole.SinglePlayer.Touch",
                    "WhiteHole.SinglePlayer.Trigger",
                    "Planet.Creation",
                    "WhiteHole.MultiPlayer.Trigger",
                    "Star.Creation",
                    "BlackHole.Creation"
                    ]
    
# Loop through trials to get meaningful_event counts over time
t = 0
for trial in trials:
    # Import raw data
    df = pd.read_json("data/logFiles/events/trial_" + str(trial) + ".log")
    # Prepare df for export
    export = pd.DataFrame()
    export["PlayTime"] = ""
    export["Trial"] = ""
    export["Position.X"] = ""
    export["Position.Y"] = ""
    export["WhiteHole.BluePlayer.Touch"] = ""
    export["WhiteHole.RedPlayer.Touch"] = ""
    export["WhiteHole.BluePlayer.Trigger"] = ""
    export["WhiteHole.RedPlayer.Trigger"] = ""
    export["Planet.BluePlayer.Creation"] = ""
    export["Planet.RedPlayer.Creation"] = ""
    export["WhiteHole.MultiPlayer.Trigger"] = ""
    export["Star.Creation"] = ""
    export["BlackHole.Creation"] = ""

    # Initiate set with 0
    export = export.append(pd.Series(0, index=export.columns), ignore_index=True)
    export.loc[0, "PlayTime"] = start_times[t]
    
    i = 1
    
    # Update value for each event
    for row in range(df.shape[0]):
        
        # Find EventType
        for event in meaningful_events:

            if (event == df.loc[row, "Event"]):
                
                # add new row with same content than last row
                export = export.append(export.loc[i-1, :], ignore_index=True)

                # Log Time Difference
                export.loc[i, "PlayTime"] = datetime.datetime.strptime(df.loc[row,"Time"], '%Y/%m/%d %H:%M:%S.%f')

                # Log Trial ID
                export.loc[i, "Trial"] = trial
                
                if (event == "WhiteHole.MultiPlayer.Trigger"):                    
                    export.loc[i, "Position.X"] = round((int(df.loc[row, "WhiteHole.1.Position.X"].split(",")[0]) + int(df.loc[row, "WhiteHole.2.Position.X"].split(",")[0]))/2)
                    export.loc[i, "Position.Y"] = round((int(df.loc[row, "WhiteHole.1.Position.Y"].split(",")[0]) + int(df.loc[row, "WhiteHole.2.Position.Y"].split(",")[0]))/2)
                    export.loc[i, event] = export.loc[i-1, event] + 1
                    
                elif (event == "Star.Creation"):
                    export.loc[i, "Position.X"] = int(df.loc[row, "Star.Position.X"].split(",")[0])
                    export.loc[i, "Position.Y"] = int(df.loc[row, "Star.Position.Y"].split(",")[0])
                    export.loc[i, event] = export.loc[i-1, event] + 1
                    
                elif (event == "BlackHole.Creation"):
                    export.loc[i, "Position.X"] = int(df.loc[row, "BlackHole.Position.X"].split(",")[0])
                    export.loc[i, "Position.Y"] = int(df.loc[row, "BlackHole.Position.Y"].split(",")[0])
                    export.loc[i, event] = export.loc[i-1, event] + 1
                    
                elif (event == "WhiteHole.SinglePlayer.Touch"):
                    export.loc[i, "Position.X"] = int(df.loc[row, "WhiteHole.Position.X"].split(",")[0])
                    export.loc[i, "Position.Y"] = int(df.loc[row, "WhiteHole.Position.Y"].split(",")[0])
                    if (df.loc[row, "Player.Color"] == "Red Player"):
                        export.loc[i, "WhiteHole.RedPlayer.Touch"] = export.loc[i-1, "WhiteHole.RedPlayer.Touch"] + 1
                    else:
                        export.loc[i, "WhiteHole.BluePlayer.Touch"] = export.loc[i-1, "WhiteHole.BluePlayer.Touch"] + 1
                    
                elif (event == "WhiteHole.SinglePlayer.Trigger"):
                    export.loc[i, "Position.X"] = int(df.loc[row, "WhiteHole.Position.X"].split(",")[0])
                    export.loc[i, "Position.Y"] = int(df.loc[row, "WhiteHole.Position.Y"].split(",")[0])
                    if (df.loc[row, "Player.Color"] == "Red Player"):
                        export.loc[i, "WhiteHole.RedPlayer.Trigger"] = export.loc[i-1, "WhiteHole.RedPlayer.Trigger"] + 1
                    else:
                        export.loc[i, "WhiteHole.BluePlayer.Trigger"] = export.loc[i-1, "WhiteHole.BluePlayer.Trigger"] + 1
                    
                elif (event == "Planet.Creation"):
                    export.loc[i, "Position.X"] = int(df.loc[row, "Planet.Position.X"].split(",")[0])
                    export.loc[i, "Position.Y"] = int(df.loc[row, "Planet.Position.Y"].split(",")[0])
                    if (df.loc[row, "Player.Color"] == "Red Player"):
                        export.loc[i, "Planet.RedPlayer.Creation"] = export.loc[i-1, "Planet.RedPlayer.Creation"] + 1
                    else:
                        export.loc[i, "Planet.BluePlayer.Creation"] = export.loc[i-1, "Planet.BluePlayer.Creation"] + 1
                i += 1
    
    # Save data in export
    export.to_csv(r'preprocessed_data/events/trial_'+ str(trial) + '.csv', index = False)
    t += 1
```

```python
# IMPORT PREPROCESSED DATA

# Re-import "compressed" event data
def importEventData(trialId):
    data = pd.read_csv("preprocessed_data/events/trial_" + str(trialId) + ".csv")
    for row in range(data.shape[0]):
        data.loc[row, "PlayTime"] = datetime.datetime.strptime(data.loc[row,"PlayTime"], '%Y-%m-%d %H:%M:%S.%f')
    return data

trials_events = []
    
for trial in trials:
    trials_events.append(importEventData(trial))

trials_events[0].head()
```

```python
# Normalize Data
for trial in trials_events:
    
    scaler = MinMaxScaler()
    
    for col in ["WhiteHole.BluePlayer.Touch", "WhiteHole.RedPlayer.Touch", "WhiteHole.BluePlayer.Trigger", "WhiteHole.RedPlayer.Trigger", "Planet.BluePlayer.Creation", "Planet.RedPlayer.Creation", "WhiteHole.MultiPlayer.Trigger", "Star.Creation", "BlackHole.Creation"]:
        x = trial[[col]].values
        trial[col] = scaler.fit_transform(x)
    
trials_events[0].head()
```
