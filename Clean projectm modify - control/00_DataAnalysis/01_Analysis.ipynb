1. `User Engagement`
2. `Behavioral Intention`
3. `Meaningful Events`
4. `User Movement`

### Data Analysis

#### Hypothesis 4: Spark leads to strong intention of its players to engage in further play sessions.

```python
# Set relevant mean
# Find the normal distributed mean
# mean_norm = statistics.mean(range(1,8))
mean_norm = 5

# Analyze the significance of the result with a 1-sample t-test
t, p = stats.ttest_1samp(bis["BIS"], mean_norm)

print("t-statistic: " + str(t))
print("p-value: " + str(p))
```

### Interpretation

H0: Spark **does not lead to a stronger intention** of its players to engage in further play sessions.

We can reject H0 at a confidence of .95 (p < .05). We can therefore assume that **Spark leads to a stronger intention to replay**.

```python
def importEventData(trialId):
    data = pd.read_csv("preprocessed_data/events/bool/trial_" + str(trialId) + ".csv")
    for row in range(data.shape[0]):
        data.loc[row, "Time"] = datetime.datetime.strptime(data.loc[row,"Time"], '%Y-%m-%d %H:%M:%S.%f')
    return data

trials_events = []
    
for trial in trials:
    trials_events.append(importEventData(trial))

trials_events[0].head()
```

### Data Analysis

```python
# Similiarity within Pairs
## Wasserstein-Metric on a X and Y positions

# intialize df to store results
wsm = pd.DataFrame()
for num in range(6):
    wsm = wsm.append(pd.Series(), ignore_index=True)

i = 0
t = 1

for trial in trials_pos:
    
    if (t==5 or t==9):
        t+=1
    
    # Split data into players
    bluePlayer = pd.DataFrame(trial.loc[trial['Player.Color'] == "Blue Player"])
    bluePlayer.pop("Time")
    bluePlayer.pop("Player.Color")
    bluePlayer.pop("Pos")
    
    redPlayer = pd.DataFrame(trial.loc[trial['Player.Color'] == "Red Player"])
    redPlayer.pop("Time")
    redPlayer.pop("Player.Color")
    redPlayer.pop("Pos")
    
    # Define cut-off for thirds:
    limit = redPlayer.shape[0]
    firstThird = round(limit/3)
    secondThird = round(limit/3*2)

    # Helpers used in for loop
    cutOff_low = [ 0, firstThird, secondThird]
    cutOff_high = [ firstThird, secondThird, limit]
    
    wsmX = [0,0,0]
    wsmY = [0,0,0]
    
    # Calculate WSM for each third of playtime for each player
    for index in [0,1,2]:       
        wsmX[index] = stats.wasserstein_distance(bluePlayer.loc[cutOff_low[index]:cutOff_high[index], "Position.X"], redPlayer.loc[cutOff_low[index]:cutOff_high[index], "Position.X"], u_weights=None, v_weights=None)
        wsmY[index] = stats.wasserstein_distance(bluePlayer.loc[cutOff_low[index]:cutOff_high[index], "Position.Y"], redPlayer.loc[cutOff_low[index]:cutOff_high[index], "Position.Y"], u_weights=None, v_weights=None)

    # Add column to df
    label = "Trial." + str(t)
    new = pd.DataFrame({label: [wsmX[0], wsmX[1], wsmX[2], wsmY[0], wsmY[1], wsmY[2]]})
    wsm = wsm.join(new)
    
    i += 1
    t += 1



# normalize data for better comparison
scaler = MinMaxScaler()
wsm  = pd.DataFrame(scaler.fit_transform(wsm))
print(wsm)
```
